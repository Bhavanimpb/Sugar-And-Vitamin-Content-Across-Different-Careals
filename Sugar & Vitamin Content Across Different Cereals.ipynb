```python
import pandas as pd
df=pd.read_csv("cereal.csv")
df
```

```python
import seaborn as sb
x=df[["sugars","vitamins"]]
sb.histplot(x)
```

```python
dic={'N': 'Nabisco',
'Q': 'Quaker Oats',
'K': 'Kelloggs',
'R': 'Raslston Purina',
'G': 'General Mills' ,
'P' :'Post' ,
'A':'American Home Foods Products'}
df["manufactures"]=[dic[x] for x in df["mfr"]]
df
```

```python
x=df.iloc[:,3:15].values
x
```

```python
y=df.iloc[:,15].values
y
```

```python
from sklearn.model_selection import train_test_split
a,b,c,d=train_test_split(x,y,test_size=0.25,random_state=0)
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(a,c)
lr_output=lr.predict(b)
lr_output
```

```python
from sklearn.metrics import mean_squared_error
mean_squared_error(lr_output,d)
```
